```python
url2="https://ci.coinafrique.com/categorie/vehicules?page=4"
```

```python
html1= requests.get(url2)
soup1=BeautifulSoup(html1.text, 'html')
print(soup)
```

```python
print(soup1.prettify())
```

```python
result2=soup1.find_all("div", class_="card ad__card round small hoverable undefined" )
```

```python
result2
```

```python

sellers = []  # Step 1: Create empty list

for nam in result2:
    seller_name= nam.find("p",class_="title")
    if seller_name:  # Make sure it's not None
        Seller= seller_name.text.strip()
        sellers.append(Seller)  # Step 2: Add to the list
        print(Seller)  
```

```python

countries= []  # Create empty list

for coun in result2:
    contry_name = coun.find("p", class_="ad__card-location")
    spani = contry_name.find("span") if contry_name else None
    
    if spani:
        Country = spani.text.strip()
        countries.append(Country)
        print(Country)

   
```

```python
descriptions = []  # Step 1: Create empty list

for des in result2:
    description = des.find("p", class_="ad__card-description monthly-payment-description")
    aa = description.find("a", class_="monthly-payment-description-text") if description else None
    Descrip = aa.text.strip() if aa else None

    descriptions.append(Descrip)  # Step 2: Add to list
    print(Descrip)  
```

```python
prices = []  # Step 1: Create empty list

for pri in result2:
    price = pri.find("p",class_="ad__card-price")  
    if price:  # Make sure it's not None
        Price = price.text.strip()
        prices.append(Price)  # Step 2: Add to the list
        print(Price)  

```

```python
descriptions1 = []  # Step 1: Create empty list

for des in result1:
    description1= des.find("p", class_="ad__card-description")
    if description1:  # Make sure it's not None
        Descrip1=description1.text.strip()
        descriptions1.append(Descrip1)  # Step 2: Add to the list
        print(Descrip1)  
        
```

```python
sellers1 = []  # Step 1: Create empty list

for nam in result1:
    seller_name1= nam.find("p",class_="title")
    if seller_name1:  # Make sure it's not None
        Seller1= seller_name1.text.strip()
        sellers1.append(Seller1)  # Step 2: Add to the list
        print(Seller1)  
```

```python
countries1= []  # Create empty list

for coun in result1:
    contry_name1 = coun.find("p", class_="ad__card-location")
    spani1 = contry_name1.find("span") if contry_name1 else None
    
    if spani1:
        Country1 = spani1.text.strip()
        countries.append(Country1)
        print(Country1)
```

```python
prices1 = []  # Step 1: Create empty list

for pri in result1:
    price1 = pri.find("p", class_="ad__card-price")
    if price1:  # Make sure it's not None
        Price1 = price1.text.strip()
        prices1.append(Price1)  # Step 2: Add to the list
        print(Price1)  
```

```python
import numpy as np
import pandas as pd
```

```python
Prices=[]
Countries=[]
Sellers= []
Descriptions = []
```

```python
Prices=prices+prices1
Countries=countries+countries1
Sellers=sellers+sellers1
Descriptions=descriptions+descriptions1
```

```python
data=pd.DataFrame(Prices,columns=["Prices"])
data1=pd.DataFrame(Countries,columns=["Countries"])
data2=pd.DataFrame(Sellers,columns=["Sellers"])
data3=pd.DataFrame(Descriptions,columns=["Descriptions"])
```

```python
df=pd.concat([data,data1,data2,data3],axis=1)
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
df=df.dropna().reset_index(drop=True)  
df.info()
```

```python
df.head(50)
```

```python
df["Prices"] = (
    df["Prices"]
    .astype(str)  # 🔄 Force everything to string
    .str.replace(r"\s+", "", regex=True)         # Remove all spaces
    .str.replace(r"[^\d]", "", regex=True)       # Remove non-digit characters (like CFA)
    .astype(int)  # 🔁 Convert cleaned result to integer
)
df.head()
```

```python
country_cleaned=[]
country_cleaned=df["Countries"].str.split(',').str[2].str.strip()
country_cleaned
```

```python
df["Countries"]=country_cleaned
df
```

```python
df1= df["Descriptions"].str.split(" ").str[0]
df1.name = "Brand"
df2 = df["Descriptions"].str.split(" ", n=1).str[1]
df2.name = "Model"
df = df.drop("Descriptions", axis=1)
```

```python
data_cheter=pd.concat([df,df1,df2],axis=1)
data_cheter
```

```python
quantity=[]
for i in range(len(data_cheter)):
    i=1
    quantity.append(i)
    print(i)
```

```python
dot=pd.DataFrame(quantity,columns=["Quantity"])
data_cheter=pd.concat([data_cheter,dot],axis=1)
data_cheter.head(100)
```

```python
data_cheter["Sellers"] = data_cheter["Sellers"].str.title()
data_cheter
```

```python
day=[]
for i in range(len(data_cheter)):
    i=18
    day.append(i)
    print(i)
```

```python
month=[]
for i in range(len(data_cheter)):
    i="Avril"
    month.append(i)
    print(i)
```

```python
year=[]
for i in range(len(data_cheter)):
    i=2025
    year.append(i)
    print(i)
```

```python

dit=pd.DataFrame(day,columns=["day"])
dat=pd.DataFrame(month,columns=["month"])
det=pd.DataFrame(year,columns=["year"])
data_cheter=pd.concat([data_cheter,dit,dat,det],axis=1)
data_cheter.head(100)
```

```python
data_cheter = data_cheter[
    ~data_cheter.apply(lambda row: row.astype(str).str.contains("NAN", case=False)).any(axis=1)
].reset_index(drop=True)
data_cheter.info()
```
